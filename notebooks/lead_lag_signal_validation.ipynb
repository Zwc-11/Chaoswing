# Lead-Lag Signal Validation

This notebook reviews mapped cross-venue pairs, candidate lead-lag signals, and paper trades so the lead-lag thesis is treated as a falsification problem rather than a claim of deployable arbitrage.

```python
from pathlib import Path
import json

import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid")

DATA_DIR = Path("../ml_data") if Path("../ml_data").exists() else Path("ml_data")

def load_jsonl(path: Path) -> pd.DataFrame:
    if not path.exists():
        return pd.DataFrame()
    with path.open("r", encoding="utf-8") as handle:
        rows = [json.loads(line) for line in handle if line.strip()]
    return pd.DataFrame(rows)

pairs_df = load_jsonl(DATA_DIR / "leadlag_pairs.jsonl")
signals_df = load_jsonl(DATA_DIR / "leadlag_signals.jsonl")
trades_df = load_jsonl(DATA_DIR / "paper_trades.jsonl")

pairs_df.head()
```

```python
summary = {
    "pairs": len(pairs_df),
    "signals": len(signals_df),
    "paper_trades": len(trades_df),
}
summary
```

```python
if not signals_df.empty:
    signal_columns = [column for column in ["signal_type", "status", "decision"] if column in signals_df.columns]
    display(signals_df[signal_columns].head())

if not trades_df.empty:
    numeric_columns = [column for column in ["gross_pnl", "net_pnl", "slippage_cost", "holding_seconds"] if column in trades_df.columns]
    if numeric_columns:
        display(trades_df[numeric_columns].describe())
```
